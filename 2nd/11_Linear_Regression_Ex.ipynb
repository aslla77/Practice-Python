```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.optimizers import SGD,Adam
import matplotlib.pyplot as plt
```

```python
x_data=np.array([[1,2,0],[5,4,3],[1,2,-1],[3,1,0],[2,4,2],
                [4,1,2],[-1,3,2],[4,3,3],[0,2,6],[2,2,1],
                [1,-2,-2],[0,1,3],[1,1,3],[0,1,4],[2,3,3]]) 
t_data=np.array([-4,4,-6,3,-4,
                 9,-7,5,6,0,
                 4,3,5,5,1])

print(x_data.shape,t_data.shape)
```

```python
model=Sequential()

model.add(Dense(1,input_shape=(3,),activation='linear'))
```

```python
model.compile(optimizer=SGD(learning_rate=1e-2),loss='mse')

model.summary()
```

```python
hist=model.fit(x_data,t_data,epochs=1000)
```

```python
test_data=([[5,5,0],[2,3,1],[-1,0,-1],[10,5,2],[4,-1,-2]])
ret_cal=[2*data[0]-3*data[1]+2*data[2] for data in test_data]

predict_val=model.predict(np.array(test_data))
print(predict_val)
print(ret_cal)
```

```python
print(model.input)
print(model.output)
print(model.weights)
```

```python
plt.title('Loss trend')
plt.xlabel('Epochs')
plt.ylabel('Loss Value')
plt.grid()
plt.plot(hist.history['loss'],label='train loss')
plt.legend(loc='best')

plt.show()
```
