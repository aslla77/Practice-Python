```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline

x_data=np.random.rand(100)
y_data=np.random.rand(100)

plt.title('scatter plot')
plt.grid()
plt.scatter(x_data,y_data,color='b',marker='o')
plt.xlim(-0.2,1.2)
plt.ylim(-0.2,1.2)
plt.show()
```

```python
x=np.arange(-5,6)
y=x**2

plt.plot(x,y)
plt.grid()
plt.xlim(-5,5)
plt.ylim(0,25)
plt.show()
```
